Create path list of dataset

```python
from glob import glob

# 이미지 경로 list로 넣기
train_img_list = glob('/content/datasets/train/images/*.jpg') + glob('/content/datasets/train/images/*.jpeg')
valid_img_list = glob('/content/datasets/valid/images/*.jpg') + glob('/content/datasets/valid/images/*.jpeg')
test_img_list = glob('/content/datasets/test/images/*.jpg') + glob('/content/datasets/test/images/*.jpeg')


# txt 파일에 write
with open('./train.txt', 'w') as f:
	f.write('\n'.join(train_img_list) + '\n')
    
with open('./valid.txt', 'w') as f:
	f.write('\n'.join(valid_img_list) + '\n')
 
with open('./test.txt', 'w') as f:
	f.write('\n'.join(valid_img_list) + '\n')
```
